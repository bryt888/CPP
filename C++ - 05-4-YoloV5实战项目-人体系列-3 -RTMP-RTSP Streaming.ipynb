### 1. How does it work

- run a "rtsp-simple-server" in one terminal
- use compiled c++ code opencv_rtmp to stream a test mp4 video file out as a stream
- use VLC as a rtmp player to receive the video from rtsp-simple-server, and play locally

**"rtsp-simple-server" is a key server. C++ program push rtmp video to this server, and VLC also try to recieve stream from this server.**

### 1.1 Run rtsp-simple server
you can find `rtsp-simple-server` from Enpei's demo code in "1.opencv_rtmp"folder


once you run the server, you will see the following lines message


### 1.2 Compile C++ code and stream out the mp4 video
- Remove the previous built "build" folder
    - `rm -fr build/`
- find the stream link in main.cpp

- compile the code
    - `cmake . -B build`
    - `cmake --build build/`


### 1.3 Use the compiled C++ code to stream out the local MP4 video file
- **Open another terminal to perform the following operations**
- `./build/opencv_rtmp ./media/test.mp4 8000000`  this test.mp4 video is a bit longer so that we will have enough time to open VLC



### 1.4 Download and setup VLC Player for RTMP 
- download the VLC Player windows version from `https://get.videolan.org/vlc/3.0.18/win64/vlc-3.0.18-win64.exe`
- go to `tools - preferences - input/codecs - network - switch to RTP over RTSP(TCP)`



- go to `media - open network stream` to enter the url`



- the url can also be found over VSCode "Ports" section


**Finally** Video shows up at VLC Player
