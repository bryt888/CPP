### 1. gflags
`sudo apt-get install libgflags-dev libgflags2.2 `

Report Error: can't find package


#### Solution:
1. update lib by: `sudo apt update`
2. run the installtion again: `sudo apt-get install libgflags-dev libgflags2.2`

`g++ main.cpp -lgflags -o main` finally works

### 2.pkg-config
`sudo apt-get install pkg-config`

`g++ main.cpp `pkg-config --cflags --libs gflags`  -o main ` works
