### 3.cuDNN 

3.1 **Download and install cuDNN**: 
- https://developer.nvidia.com/rdp/cudnn-download

- **Install tcuDNN after downloaded**
    - https://docs.nvidia.com/deeplearning/cudnn/install-guide/index.html



### 3.2 Install cuDNN

#### 注意，运行下面的命令前，将下面的 X.Y和v8.x.x.x 替换成自己具体的CUDA 和 cuDNN版本，

- my CUDA: 11.8
- cuDNN:8.8.1.3

- `sudo dpkg -i cudnn-local-repo-${OS}-8.x.x.x_1.0-1_amd64.deb`
    - my version: `sudo dpkg -i cudnn-local-repo-ubuntu2004-8.8.1.3_1.0-1_amd64.deb`

- sudo cp /var/cudnn-local-repo-*/cudnn-local-*-keyring.gpg /usr/share/keyrings/

- sudo apt-get update

- **Install cudnn:**
    - sudo `apt-get install libcudnn8=8.x.x.x-1+cudaX.Y`
    - my version：`sudo apt-get install libcudnn8=8.8.1.3-1+cuda11.8`

- **Install cudnn-dev**
    - `sudo apt-get install libcudnn8-dev=8.x.x.x-1+cudaX.Y`
    - my version: `sudo apt-get install libcudnn8-dev=8.8.1.3-1+cuda11.8`

- **Install cudnn-samples:**
    - `sudo apt-get install libcudnn8-samples=8.x.x.x-1+cudaX.Y`
    - my version：`sudo apt-get install libcudnn8-samples=8.8.1.3+cuda11.8`

### 3.3 Verify Cudnn installation
- `cp -r /usr/src/cudnn_samples_v8/ $HOME`
- `cd  $HOME/cudnn_samples_v8/mnistCUDNN`
- `make clean && make`

**Had Issue: FreeImage.h can't be found**


**Fixed by the following Installation:**<br>
`sudo apt-get install libfreeimage3 libfreeimage-dev`

- `./mnistCUDNN`

### Passed!!!

### 4. TensorRT

### 4.1 Download driver
https://developer.nvidia.com/nvidia-tensorrt-8x-download

- my version for ubuntu22.04: https://developer.nvidia.com/downloads/compute/machine-learning/tensorrt/secure/8.5.3/local_repos/nv-tensorrt-local-repo-ubuntu2204-8.5.3-cuda-11.8_1.0-1_amd64.deb

### 4.2 Install TensorRT
Reference: https://docs.nvidia.com/deeplearning/tensorrt/install-guide/index.html#installing-debian

- Replace variables below to your own version
    - os="ubuntuxx04"
    - tag="8.x.x-cuda-x.x"
    
- **Install package:**    
    - `sudo dpkg -i nv-tensorrt-local-repo-${os}-${tag}_1.0-1_amd64.deb`
    - my version：`sudo dpkg -i nv-tensorrt-local-repo-ubuntu2004-8.5.3-cuda-11.8_1.0-1_amd64.deb`
- **copy key:**`
    - `sudo cp /var/nv-tensorrt-local-repo-${os}-${tag}/*-keyring.gpg /usr/share/keyrings/`
    - my version：`sudo cp /var/nv-tensorrt-local-repo-ubuntu2004-8.5.3-cuda-11.8/*-keyring.gpg /usr/share/keyrings/`

- **`sudo apt-get update`**
- **`sudo apt-get install tensorrt`**

### 4.3 Verification
- **`dpkg -l | grep TensorRT`**
- **`dpkg-query -W tensorrt`**



Remove TensorRT

`sudo apt-get remove tensorrt`
