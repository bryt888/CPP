### 1. Prepare YoloV5 in Docker Image

- Clone project: `git clone https://github.com/ultralytics/yolov5.git`
- Enter folder: `cd yolov5`
- check out specified version to avoid compatible problem `git checkout a80dd66efe0bc7fe3772f259260d5b7278aab42f`
- install required libraries and files in docker environment:  `pip3 install -r requirements.txt  `


### Download the pre-trained weight 
wget https://github.com/ultralytics/yolov5/releases/download/v7.0/yolov5s.pt

docker容器内部可能报错：AttributeError: partially initialized module 'cv2' has no attribute '_registerMatType' (most likely due to a circular import)，使用pip3 install "opencv-python-headless<4.3"

### 2. Test Yolov5
`python detect.py --source ./data/images/ --weights yolov5s.pt --conf-thres 0.4`



### 3. Prepare Training Data
